```python
import numpy as np
import pandas as pd
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">1.2 - Importation des fichiers</h3>
</div>

```python
# Importation du fichier aide_alimentaire 
aide_alimentaire=pd.read_csv('aide_alimentaire2.csv')
aide_alimentaire.rename(columns={"Pays bénéficiaire":"Zone"},inplace=True) # Renommage colonne Pays bénéficiaire par Zone
aide_alimentaire.head()
```

```python
# Importation du fichier dispo_alimentaire
dispo_alimentaire=pd.read_csv('dispo_alimentaire2.csv')
dispo_alimentaire.head()
```

```python
# Importation du fichier population
population=pd.read_csv('population2.csv')
population.rename(columns={"Année":"Année pop", "Valeur":"Population"},inplace=True) # Renommage des colonnes Année et Valeur
population["Population"] = round(population["Population"]/1000,2) # Population en millions d'habitants
population.head()
```

```python
# Filtre sur l'année 2017
population_2017=population.loc[population['Année pop']==2017] 
population_2017.head()
```

```python
# Importation du fichier sous_nutrition
sous_nutrition=pd.read_csv('sous_nutrition2.csv')
sous_nutrition.rename(columns={"Année":"Année sous nutrition", "Valeur":"Population sous nutrition"},inplace=True) # Renommage des colonnes Année et Valeur
sous_nutrition.replace("<0.1","0.1", inplace = True) # Remplacement des valeurs inférieures à 0.1 par 0.1
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].str.replace('2012-2014', '2013') # Remplacement par la moyenne de la fourchette des années
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].str.replace('2013-2015', '2014') # Remplacement par la moyenne de la fourchette des années
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].str.replace('2014-2016', '2015') # Remplacement par la moyenne de la fourchette des années
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].str.replace('2015-2017', '2016') # Remplacement par la moyenne de la fourchette des années
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].str.replace('2016-2018', '2017') # Remplacement par la moyenne de la fourchette des années
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].str.replace('2017-2019', '2018') # Remplacement par la moyenne de la fourchette des années
sous_nutrition['Population sous nutrition']= pd.to_numeric(sous_nutrition ['Population sous nutrition'],errors = 'coerce')# Modification du type de données
sous_nutrition['Année sous nutrition'] = sous_nutrition['Année sous nutrition'].astype(int) # Modification du type de données
sous_nutrition.head()
```

```python
# Filtre sur l'année 2017
sous_nutrition_2017=sous_nutrition.loc[sous_nutrition['Année sous nutrition']==2017]
sous_nutrition_2017.head()
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">2.1 - Calcul de la proportion de personnes en état de sous-nutrition</h3>
</div>

```python
# Fusion des fichiers population & sous_nutrition sur année 2017 
df_population_sous_nutrition_2017=pd.merge(population_2017,sous_nutrition_2017,left_on=['Zone', 'Année pop'],
                                           right_on=['Zone', 'Année sous nutrition'],how='left')
df_population_sous_nutrition_2017.head()
```

```python
# Calcul de la proportion de personnes
proportion_mondiale_2017 = round(df_population_sous_nutrition_2017['Population sous nutrition'].sum() 
                                 / df_population_sous_nutrition_2017['Population'].sum()*100,2)
print("La proportion de la population mondiale en état de sous-nutrition en 2017 est de : ", proportion_mondiale_2017, "%")
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">2.2 - Calcul du nombre théorique de personnes qui pourraient être nourries</h3>
</div>

```python
# Sélection des données pour calcul théorique de personnes qui pourraient être nourries
dispo_alimentaire_theorique=dispo_alimentaire.groupby('Zone')['Disponibilité alimentaire (Kcal/personne/jour)'].sum()
dispo_alimentaire_theorique.head()
```

```python
# Fusion des fichiers dispo_par_zone & population
dispo_ali_pays_2017 = pd.merge(dispo_alimentaire_theorique, population_2017, on='Zone')
population_totale= round(dispo_ali_pays_2017['Population'].sum(),2)
dispo_ali_pays_2017.head()
```

```python
# Création d'une nouvelle colonne disponibilité alimentaire totale
dispo_ali_pays_2017['Dispo_ali_totale']=round(dispo_ali_pays_2017['Disponibilité alimentaire (Kcal/personne/jour)']*
                                              dispo_ali_pays_2017['Population'],2)
dispo_ali_pays_2017.head()
```

```python
# Calcul du nombre théorique de personnes qui pourraient être nourries 
nombre_théorique = round(dispo_ali_pays_2017['Dispo_ali_totale'].sum () / 2500,2)
taux_population=round((nombre_théorique/population_totale),2) *100
print("Le nombre théorique de personnes qui pourraient être nourries est de", round(nombre_théorique/1000,2),
      "milliards", "soit",round(taux_population,2), "% de la population.")
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">2.3 - Calcul du nombre théorique de personnes qui pourraient être nourries avec des produits végétaux</h3>
</div>

```python
# Sélection des données pour calcul théorique de personnes qui pourraient être nourries de produits végétaux
dispo_alimentaire_theorique_vegetale=dispo_alimentaire[['Zone','Origine','Disponibilité alimentaire (Kcal/personne/jour)']]
dispo_alimentaire_theorique_vegetale = dispo_alimentaire_theorique_vegetale.loc[dispo_alimentaire_theorique_vegetale['Origine']
                                                                                == 'vegetale'] # Filtre sur les produits végétaux
dispo_alimentaire_theorique_vegetale.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)':'Dispo ali végétale'}, 
                                            inplace = True) # Renommage de la colonne Disponibilité alimentaire
dispo_alimentaire_theorique_vegetale=dispo_alimentaire_theorique_vegetale.groupby('Zone')['Dispo ali végétale'].sum()
dispo_alimentaire_theorique_vegetale.head()
```

```python
# Fusion des fichiers dispo_par_zone & population
dispo_vegetale_pays_2017 = pd.merge(dispo_alimentaire_theorique_vegetale, population_2017, on='Zone')
dispo_vegetale_pays_2017.head()
```

```python
# Création d'une nouvelle colonne Dispo_vegetale_totale
dispo_vegetale_pays_2017['Dispo_vegetale_totale']=dispo_vegetale_pays_2017['Dispo ali végétale']*dispo_vegetale_pays_2017['Population']
dispo_vegetale_pays_2017.head()
```

```python
# Calcul du nombre théorique de personnes qui pourraient être nourries par des produits végétaux
nombre_théorique2 = round(dispo_vegetale_pays_2017['Dispo_vegetale_totale'].sum() / 2500 ,2)
taux_population2=round((nombre_théorique2/population_totale),2)*100
print("Le nombre théorique de personnes qui pourraient être nourries de produits végétaux est de", 
      round(nombre_théorique2/1000,2), "milliards","soit",round(taux_population2,2),"% de la population")
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">2.4 - Disponibilité intérieure utilisée, part attribuée à l'alimentation animale, celle attribuée à l'alimentation humaine et celle perdue</h3>
</div>

```python
# Calcul de Julien : Aliments pour animaux + Autres Utilisations + Nourriture + Pertes + Semences + Traitement = 
# disponibilité intérieure = Production + Variation de stock + Importation - Exportation
# Création du nouveau dataframe disponibilité intérieure

disponibilité_intérieure = dispo_alimentaire.groupby(by=['Zone'])[['Produit', 'Origine', 'Aliments pour animaux','Autres Utilisations','Disponibilité intérieure',
                                                                   'Exportations - Quantité','Importations - Quantité', 
                                                                   'Nourriture','Pertes','Production', 'Semences', 'Traitement',
                                                                   'Variation de stock']].sum()
disponibilité_intérieure.head()
```

* Création de 3 nouvelles colonnes : 
 - Disponibilité intérieure utilisée pour l'alimentation animale en %
 - Disponibilité intérieure utilisée pour l'alimentation humaine en %
 - Disponibilité intérieure perdue en %

```python
disponibilité_intérieure['Disponibilité intérieure pour les animaux en %'] = round(disponibilité_intérieure['Aliments pour animaux'] / disponibilité_intérieure['Disponibilité intérieure']*100,2)
disponibilité_intérieure['Disponibilité intérieure pour les humains en %'] = round(disponibilité_intérieure['Nourriture'] / disponibilité_intérieure['Disponibilité intérieure']*100,2)
disponibilité_intérieure['Disponibilité intérieure perdue en %'] = round(disponibilité_intérieure['Pertes']/disponibilité_intérieure['Disponibilité intérieure']*100,2)
disponibilité_intérieure.head()
```

```python
# Calcul de la disponibilité intérieure moyenne
disponibilité_intérieure2=disponibilité_intérieure[['Disponibilité intérieure','Disponibilité intérieure pour les animaux en %','Disponibilité intérieure pour les humains en %','Disponibilité intérieure perdue en %']].mean()
disponibilité_intérieure2.head()
```

La disponibilité intérieure moyenne en 2017 était de 56 603,41 milliers de tonnes, la part réservée pour les animaux était de 12,02 %, celle pour les humains, de 57,52 % et celle perdue, 4,66 %

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.1 - Utilisation des céréales : répartition entre l'alimentation humaine et l'alimentation pour animaux</h3>
</div>

* Filtre sur les céréales et création de 2 nouvelles colonnes :
 - Nourriture en %
 - Aliments pour animaux en %

```python
liste_cereales = ['Blé', 'Céréales Autres','Maïs','Millet','Orge','Riz (Eq Blanchi)','Avoine','Sorgho','Seigle']
dispo_alimentaire_cereales = dispo_alimentaire.loc[dispo_alimentaire['Produit'].isin(liste_cereales)]
dispo_alimentaire_cereales=dispo_alimentaire_cereales.groupby(by=['Zone'])[['Produit','Disponibilité intérieure',
                                                                            'Aliments pour animaux','Nourriture']].sum()
dispo_alimentaire_cereales['Nourriture en %'] = round(dispo_alimentaire_cereales['Nourriture'] / 
                                                      dispo_alimentaire_cereales['Disponibilité intérieure']*100,2)
dispo_alimentaire_cereales['Aliments pour animaux en %'] = round(dispo_alimentaire_cereales['Aliments pour animaux']/
                                                                 dispo_alimentaire_cereales['Disponibilité intérieure']*100,2)
dispo_alimentaire_cereales.head()
```

```python
# Calcul de la disponibilité intérieure moyenne
dispo_alimentaire_cereales=dispo_alimentaire_cereales[['Disponibilité intérieure','Aliments pour animaux','Nourriture',
                                                       'Nourriture en %','Aliments pour animaux en %']].mean()
dispo_alimentaire_cereales.head()
```

La disponibilité intérieure moyenne de céréales en 2017 était de 13 833,33 milliers de tonnes, la part réservée pour les animaux était de 29,15 %, celle pour les humains, de 52,97 %.

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Utilisation du manioc aux égards de la proportion de personnes en sous-nutrition </h3>

* Fusion des fichiers dispo_par_zone et sous_nutrition et création de 3 nouvelles colonnes :
 - taux_exportations
 - taux_importations
 - taux_nourriture

```python
thailande = pd.merge(dispo_alimentaire, sous_nutrition_2017, on='Zone')
thailande_manioc=thailande.loc[(thailande['Produit']=='Manioc')&(thailande['Zone']=='Thaïlande')]
thailande_manioc=thailande_manioc[['Zone','Produit','Exportations - Quantité','Importations - Quantité','Production',
                                   'Nourriture','Population sous nutrition']]
thailande_manioc['taux_exportations']=round(thailande_manioc['Exportations - Quantité']/thailande_manioc['Production']*100,2)
thailande_manioc['taux_importations']=round(thailande_manioc['Importations - Quantité']/thailande_manioc['Production']*100,2)
thailande_manioc['taux_nourriture']=round(thailande_manioc['Nourriture']/thailande_manioc['Production']*100,2)
thailande_manioc.head() 
```

En 2017, la population de la Thaïlande en sous-nutrition était de 6,20 millions d'habitants, sa production de Manioc était de 30 228 milliers de tonnes, ses exportations représentaient 83,41 % de sa production alors que ses importations étaient de 4,14 %, la part consacrée à la nourriture n'était que de 2,88 %.
